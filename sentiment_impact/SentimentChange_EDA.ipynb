```python
import numpy as np
import pandas as pd
from tqdm import tqdm_notebook
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('conversations_with_sentimpact.csv', encoding='ISO-8859-1')
```

```python
df2 = df[df.astype(str)['SentChange'] != '[]']
all_sentchanges = []
for data in df2[['SentChange']].values:
    sentchange = eval(data[0])
    for item in sentchange:
        all_sentchanges.append(item)
        
print(len(all_sentchanges))
```

```python
Changing_Sentiment = pd.Series(all_sentchanges)
```

```python
Changing_Sentiment.describe()
```

```python
Changing_Sentiment.hist(bins=5, color='blue')
mean = Changing_Sentiment.mean()
std = Changing_Sentiment.std()
plt.axvline(x=mean, linewidth=5, color='red')
plt.text(x=mean-1.3, y = 25000, s='Mean: ' + str(round(mean,3)) + '\nStd. Dev. : ' + str(round(std, 3)), color='red', size=14)
plt.savefig('Distribution_sentiment_change.jpg')
plt.title('Distribution of Sentiment Change', size=14)
plt.xlabel('Sentiment Change', size=12)
plt.ylabel('Density',size=12)

plt.show()
```
