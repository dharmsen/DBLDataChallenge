```python
import sqlite3
import pandas as pd
```

```python
def update_db_conversations(fp_db, df_convo):
    conn = sqlite3.connect(fp_db)
    df_convo.to_sql('conversations', conn, index="False", if_exists="replace")
    conn.close()

    print('Done!')
    
```

```python
df_wrangled = pd.read_csv("ALL_CONVERSATIONS_WRANGLED.csv")
df_text = pd.read_csv("CONVERSATIONS_TEXT.csv")
df_wrangled['full_text'] = df_text['Conversations']
```

```python
update_db_conversations("ALL_DATA_32_FEATURES.db", df_wrangled)
```
