```python
import re
import sqlite3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook
from nltk.tokenize import word_tokenize

conn = sqlite3.connect('full_conversation_database.db')
df = pd.read_sql('SELECT * from tweets', conn)
df.head(2)
```

```python
KLM_incoming = [text for text in list(df['text']) if '@KLM' in text]
```

```python
pattern = re.compile('^[0-9]{2}[A-Z]{4}$')
pattern2 = re.compile('^[0-9]{1}[A-Z]{5}$')
#pattern3 = re.compile('^[0-9]{2}[a-z]{4}$')
#pattern4 = re.compile('^[0-9]{1}[a-z]{5}$')
tweets = []
book_code = []
for t in tqdm_notebook(KLM_incoming):
    for word in word_tokenize(t):
        if pattern.match(word) or pattern2.match(word):# or pattern3.match(word) or pattern4.match(word):
            #if 'mins' != word[-4:] or 'euro' != word[-4:] or 'days' != word[-4:] or 'dagen' != word[-5:] or 'delay' != word[-5:] or 'hours' != word[-5:]:
            tweets.append(t)
            book_code.append(word)
            #else:
             #   pass
```

```python
print(f'Booking codes collected: {len(book_code)}')
print(f'Unique booking codes collected: {len(set(book_code))}')
```
