```python
import pandas as pd
import sqlite3
from DataBaseInterface import LoadDatabaseAsDF

conn = sqlite3.connect('ALL_DATA_8_FEATURES.db')
created_df = pd.read_sql("SELECT * FROM conversations", conn)
tweets_df = pd.read_sql("SELECT * FROM tweets", conn)
```

```python
KLM_conversations = created_df[created_df['airlines_involved'] == "['KLM']"]
BA_conversations = created_df[created_df['airlines_involved'] == "['British_Airways']"]
lufthansa_conversations = created_df[created_df['airlines_involved'] == "['Lufthansa']"]
ryanair_conversations = created_df[created_df['airlines_involved'] == "['RyanAir']"]
easyjet_conversations = created_df[created_df['airlines_involved'] == "['easyJet']"]
singaporeair_conversations = created_df[created_df['airlines_involved'] == "['SingaporeAir']"]
qantas_conversations = created_df[created_df['airlines_involved'] == "['Qantas']"]
virginatlantic_conversations = created_df[created_df['airlines_involved'] == "['VirginAtlantic']"]
airfrance_conversations = created_df[created_df['airlines_involved'] == "['AirFrance']"]
etihad_conversations = created_df[created_df['airlines_involved'] == "['EtihadAirways']"]
american_conversations = created_df[created_df['airlines_involved'] == "['AmericanAir']"]
```

```python
airline_convo_length = pd.DataFrame({'airline': ["KLM", "British Airways", "Lufthansa", "RyanAir", "easyJet", "SingaporeAir", "Qantas", "Virgin Atlantic", "AirFrance", "Etihad Airways", "American Air"], "avg length": [KLM_conversations['conversation_length'].mean(), BA_conversations['conversation_length'].mean(), lufthansa_conversations['conversation_length'].mean(), ryanair_conversations['conversation_length'].mean(), easyjet_conversations['conversation_length'].mean(), singaporeair_conversations['conversation_length'].mean(), qantas_conversations['conversation_length'].mean(), virginatlantic_conversations['conversation_length'].mean(), airfrance_conversations['conversation_length'].mean(), etihad_conversations['conversation_length'].mean(), american_conversations['conversation_length'].mean()]})
airline_convo_length = airline_convo_length.set_index('airline')
airline_convo_length = airline_convo_length.sort_values('avg length', ascending=False)
```

```python
KLM_incoming_hours = []
BA_incoming_hours = []
AA_incoming_hours = []
L_incoming_hours = []
EJ_incoming_hours = []
RA_incoming_hours = []
SA_incoming_hours = []
Q_incoming_hours = []
EA_incoming_hours = []
VA_incoming_hours = []
AF_incoming_hours = []

for i in tweets_df['text']:
    if '@KLM' in i:
        KLM_incoming_hours.append(i)
    elif '@British_Airways' in i:
        BA_incoming_hours.append(i)
    elif '@AmericanAir' in i:
        AA_incoming_hours.append(i)
    elif '@Lufthansa' in i:
        L_incoming_hours.append(i)
    elif '@easyJet' in i:
        EJ_incoming_hours.append(i)
    elif '@RyanAir' in i:
        RA_incoming_hours.append(i)
    elif '@SingaporeAir' in i:
        SA_incoming_hours.append(i)
    elif '@Qantas' in i:
        Q_incoming_hours.append(i)
    elif '@EtihadAirways' in i:
        EA_incoming_hours.append(i)
    elif '@VirginAtlantic' in i:
        VA_incoming_hours.append(i)
    elif '@AirFrance' in i:
        AF_incoming_hours.append(i)
        
KLM_incoming = len(pd.Series(KLM_incoming_hours))
BA_incoming = len(pd.Series(BA_incoming_hours))
AA_incoming = len(pd.Series(AA_incoming_hours))
L_incoming = len(pd.Series(L_incoming_hours))
EJ_incoming = len(pd.Series(EJ_incoming_hours))
RA_incoming = len(pd.Series(RA_incoming_hours))
SA_incoming = len(pd.Series(SA_incoming_hours))
Q_incoming = len(pd.Series(Q_incoming_hours))
EA_incoming = len(pd.Series(EA_incoming_hours))
VA_incoming = len(pd.Series(VA_incoming_hours))
AF_incoming = len(pd.Series(AF_incoming_hours))
```

```python
KLM_tweet_volume = KLM_incoming + KLM_outgoing
BA_tweet_volume = BA_incoming + BA_outgoing
AA_tweet_volume = AA_incoming + american_outgoing
L_tweet_volume = L_incoming + lufthansa_outgoing
EJ_tweet_volume = EJ_incoming + easyjet_outgoing
RA_tweet_volume = RA_incoming + ryanair_outgoing
SA_tweet_volume = SA_incoming + singaporeair_outgoing
Q_tweet_volume = Q_incoming + qantas_outgoing
EA_tweet_volume = EA_incoming + etihad_outgoing
VA_tweet_volume = VA_incoming + virginatlantic_outgoing
AF_tweet_volume = AF_incoming + airfrance_outgoing
```
