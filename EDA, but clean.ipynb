```python
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
import seaborn as sns

sys.path.insert(0, '/data_extraction/')
from DataBaseInterface import LoadDatabaseAsDF

conn = sqlite3.connect('ALL_DATA_32_FEATURES.db')
created_df = pd.read_sql("SELECT [('user', 'id_str')], created_at FROM tweets", conn)

KLM_tweets = created_df[created_df["('user', 'id_str')"] == "56377143"].reset_index()
BA_tweets = created_df[created_df["('user', 'id_str')"] == "18332190"].reset_index()
```

```python
dates_british = []
dates_KLM = []
month = 0

for i in BA_tweets['created_at']:
    if i[4:7] == "Jan":
        month = 1
    elif i[4:7] == "Feb":
        month = 2
    elif i[4:7] == "Mar":
        month = 3
    elif i[4:7] == "Apr":
        month = 4
    elif i[4:7] == "May":
        month = 5
    elif i[4:7] == "Jun":
        month = 6
    elif i[4:7] == "Jul":
        month = 7
    elif i[4:7] == "Aug":
        month = 8
    elif i[4:7] == "Sep":
        month = 9
    elif i[4:7] == "Oct":
        month = 10
    elif i[4:7] == "Nov":
        month = 11
    elif i[4:7] == "Dec":
        month = 12

    dates_british.append({'year': int(i[26:30]), 'month': month, 'day': int(i[8:10])})
    
for i in KLM_tweets['created_at']:
    if i[4:7] == "Jan":
        month = 1
    elif i[4:7] == "Feb":
        month = 2
    elif i[4:7] == "Mar":
        month = 3
    elif i[4:7] == "Apr":
        month = 4
    elif i[4:7] == "May":
        month = 5
    elif i[4:7] == "Jun":
        month = 6
    elif i[4:7] == "Jul":
        month = 7
    elif i[4:7] == "Aug":
        month = 8
    elif i[4:7] == "Sep":
        month = 9
    elif i[4:7] == "Oct":
        month = 10
    elif i[4:7] == "Nov":
        month = 11
    elif i[4:7] == "Dec":
        month = 12

    dates_KLM.append({'year': int(i[26:30]), 'month': month, 'day': int(i[8:10])})    
```

```python
# on what day of the week are the airlines most likely to respond?

import datetime
weekdays_BA = []
weekdays_KLM = []

for i in dates_british:
    weekdays_BA.append(datetime.date(i['year'], i['month'], i['day']).weekday())
    
for i in dates_KLM:
    weekdays_KLM.append(datetime.date(i['year'], i['month'], i['day']).weekday())
```

```python
# amt of tweets each weekday

from collections import Counter

num_weekdays_KLM = Counter(weekdays_KLM).values()
num_weekdays_BA = Counter(weekdays_BA).values()
```

```python
time_created_BA = BA_tweets['created_at']
time_created_int = time_created_BA.str[11:13].astype(int)
BA_hour_counts = time_created_int.value_counts().sort_index()

time_created_KLM = KLM_tweets['created_at']
time_created_KLM_int = time_created_KLM.str[11:13].astype(int)
KLM_hour_counts = time_created_KLM_int.value_counts().sort_index()

result = pd.concat([BA_hour_counts, KLM_hour_counts], axis=1)

sns.set()
color = sns.color_palette("colorblind")
result.plot(figsize=(10, 5))
plt.title('Twitter Activity by Hour of Day', weight='bold', fontsize=20)
plt.xlabel('Hour (GMT)', fontsize=16)
plt.ylabel('Tweets', fontsize=16);
plt.legend(['BA', 'KLM'])
plt.savefig('time of day.png', dpi=300)
```

```python
dates_british_df = pd.DataFrame(columns=['year', 'month', 'day'])

for i in dates_british:
    dates_british_df.loc[len(dates_british_df)] = i
british_records = pd.DataFrame()
british_records = dates_british_df.groupby(dates_british_df.columns.tolist()).size().reset_index().\
    rename(columns={0:'records'})
british_records
```

```python
fig, ax = plt.subplots(figsize=(10,5))
british_records = british_records.pivot("month", "day", "records")
plot = sns.heatmap(british_records)
plot.set_title("British Airways Tweets Per Day")
plt.savefig('year_overview_BA', dpi=300)
```
